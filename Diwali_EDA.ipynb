```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Deleting Status and unnamed1 column since it doesn't has any records
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#deleting na from Amount since it has very less missing values
df.dropna(how='any',inplace=True)
```

```python
#changing dtype to int from float
df['Amount']=df['Amount'].astype('int')
```

```python
df[['Age','Orders','Amount']].describe()
```

# Visualizations

## Gender column

```python
df.columns
```

```python
ax = sns.countplot(x='Gender',data=df,width=0.2)
for bars in ax.containers:
    ax.bar_label(bars)
```

- Female customers are almost twice than Male customers

# Age Group column

```python
ax = sns.countplot(x='Age Group',data =df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

- Female and males in age group 26-35 has more pusrchasing power

```python
ax = sns.countplot(x='Age Group',data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

- Age group 26-36 has more purchasing power

```python
sns.histplot(x='Age Group',data =df,hue='Gender',fill=True)
```

# State column

```python
df.columns
```

```python
 df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.title("Total order placed by States")
```

- More orders was placed from Uttar Pradesh followed by Maharashtra and Karnataka

```python
df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
amount_states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=amount_states)
plt.title("Total Amount spent by States")
```

## Marital_Status

```python
df.Marital_Status.value_counts()
```

```python
df.groupby(['Marital_Status','Gender'],as_index=False)['Orders'].sum()
```

```python
marital_orders = df.groupby(['Marital_Status','Gender'],as_index=False)['Orders'].sum()
```

```python
ax = sns.barplot(x='Marital_Status',y='Orders',data=marital_orders,hue = 'Gender')
plt.title("Orders By Marital_status")
for bars in ax.containers:
    ax.bar_label(bars)
```

- Unmarried Female and males are likely to buy more products

## Zone column

```python
df.Zone.value_counts()
```

```python
sns.countplot(y='Zone',order=df['Zone'].value_counts().index[0:5],data =df)
```

- more purchase is done from central zone

## Product_Category

```python
plt.figure(figsize=(25,5))
sales_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)
sns.barplot(data=sales_cat,x='Product_Category',y='Amount')
```

- Top amount spent categories are Food, Clothing & Apparel and Electronics & Gadgets

```python
df.Product_Category.value_counts()
```

```python
sns.countplot(y='Product_Category',order= df['Product_Category'].value_counts().index[0:10],data=df)
```

- Top 3 product categories are Clothes & Apparel, Food and Electronic & Gadgets

## Ocupation Column

```python
df.Occupation.value_counts()
```

```python
Occupation_Gender_orders = df.groupby(['Occupation','Gender'],as_index=False)['Orders'].sum()
```

```python
plt.figure(figsize=(25,5))
sns.barplot(x='Occupation',y='Orders',data=Occupation_Gender_orders,hue='Gender')
```

- Females of IT Sector, Healthcare and Aviation Occupations place more orders

## Product_ID Column

```python
df.Product_ID.value_counts().head(10)
```

```python
sns.countplot(x='Product_ID',order =df['Product_ID'].value_counts().index[0:10] ,data=df)
```

```python
Pid_orders= df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
ax = sns.barplot(x='Product_ID',y='Orders',data =Pid_orders)
for bars in ax.containers:
    ax.bar_label(bars)
```

- Mostly saled product is P00265242
